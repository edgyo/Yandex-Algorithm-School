## Сложность алгоритмов

**Сложность алгоритма** - порядок количества действий, которые выполняет алгоритма

1. O(n) - программа с одним циклом
2. O(n^2) - программа с двумя вложенными циклами
3. 100\*N = O(n) и 2\*N = O(n) - алгоритмами с констатами.
4. Пространственная сложность

**set()** - делает из многих структур питона множество встречающихся значений

Задача: найти, сколько раз встречается каждый символ в строке

Решение: Пройтись по каждой уникальной букве в слове и посчитать кол-во раз.

Оптимальное решение: завести словарь, где ключом является символ, а значением - сколько раз он встретился.

```python
# Полуоптимизированное решение

s = input()
ans = ''
anscnt = 0
for now in set(s):
    nowcnt = 0
    for j in range(len(s)):
        if now == s[j]:
            nowcnt += 1
    if nowcnt > anscnt:
        ans = now
        anscnt = nowcnt
print(ans)
```

```python
# Оптимизированное решение
s = input()
dict = {}
ans = ''
anscnt = 0
for now in s:
    if now not in dict:
        dict[now] = 0
    dict[now] += 1

for key in dict:
    if dict[key] > anscnt:
        anscnt = dict[key]
        ans = key

print(ans, anscnt)
```
